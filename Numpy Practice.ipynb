```python
#基於位置列表的索引方法
```

```python
import scipy.misc
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the Lena array
lena = scipy.misc.lena()
xmax = lena.shape[0]
ymax = lena.shape[1]
```

```python
def shuffle_indices(size):
   arr = np.arange(size)
   np.random.shuffle(arr)   #將數組隨機打亂

   return arr
```

```python
print shuffle_indices(10)
```

```python

xindices = shuffle_indices(xmax)
np.testing.assert_equal(len(xindices), xmax)
yindices = shuffle_indices(ymax)
np.testing.assert_equal(len(yindices), ymax)
```

```python
# Plot Lena
plt.imshow(lena[np.ix_(xindices, yindices)])
plt.show()
```

```python
#ix 函數  可以將傳進去多數組 會回傳最後一個數組被數組包起來  以及前面的數組會變成一個一個獨立的數組在用數組包起來
#ex
np.ix_([0,1],[2,3],[3,4])
#print np.ix_(xindices, yindices)[0]
#lena[np.ix_(xindices, yindices)]
```

```python
lena[np.ix_(xindices, yindices)]  #可以選出一個數組中有n獨立數組 以及打亂的一個完整數組
```

```python
#用布爾型變量索引Numpy 數組  boolean indexing
```

```python
import scipy.misc
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the Lena array
lena = scipy.misc.lena()
```

```python
def get_indices(size):
   arr = np.arange(size)
   return arr % 4 == 0
```

```python
#test get_indices()  
get_indices(20)   #回傳一個numpy array 根據傳入size  產生  並且每四個第一個會是True False
```

```python
# Plot Lena
lena1 = lena.copy()   #產生一個lena copy
xindices = get_indices(lena.shape[0])  #取得指定的x軸陣列 以True False False False True...
#print xindices
yindices = get_indices(lena.shape[1])
#print yindices
lena1[xindices, yindices] = 0  #透過True Fasle 將True 的資料變為0
print lena1[5,5]
```

```python
plt.subplot(211)
plt.imshow(lena1)
```

```python
lena2 = lena.copy() 
# Between quarter and 3 quarters of the max value
lena2[(lena > lena.max()/4) & (lena < 3 * lena.max()/4)] = 0  #針對大於最大值的1/4 以及3/4的資料 將其置於0

plt.subplot(212)
plt.imshow(lena2)


plt.show()
```

```python
import scipy.io.wavfile
import matplotlib.pyplot as plt
import urllib2
import numpy as np
```

```python
response = urllib2.urlopen('http://www.thesoundarchive.com/austinpowers/smashingbaby.wav') #透過response 取得下載的回應
print response.info()
```

```python
WAV_FILE = 'smashingbaby.wav'
filehandle = open(WAV_FILE, 'w')
filehandle.write(response.read())  #read 可以獲得前面抓取的網頁內容
filehandle.close()
sample_rate, data = scipy.io.wavfile.read(WAV_FILE)  #回傳數據及採樣率
print "Data type", data.dtype, "Shape", data.shape
```

```python
plt.subplot(2, 1, 1)
plt.title("Original")
plt.plot(data)
```

```python
newdata = data * 0.2  
newdata = newdata.astype(np.uint8)
print "Data type", newdata.dtype, "Shape", newdata.shape
```

```python
scipy.io.wavfile.write("quiet.wav",
    sample_rate, newdata)
```

```python
plt.subplot(2, 1, 2)
plt.title("Quiet")
plt.plot(newdata)

plt.show()
```
